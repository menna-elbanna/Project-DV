```python
import pandas as pd
import plotly.express as px
from google.colab import files
uploaded = files.upload()
df = pd.read_csv("/content/clean_players.csv")

print("Columns in dataset:")
print(df.columns)

print("\nSample data:")
print(df.head())

club_col = None
pos_col = None

for col in df.columns:
    if 'club' in col.lower():
        club_col = col
    if 'position' in col.lower():
        pos_col = col

print(f"\nDetected Club Column: {club_col}")
print(f"Detected Position Column: {pos_col}")

df['position_main'] = df[pos_col].astype(str).str.split(',').str[0]

df = df.dropna(subset=[club_col, 'position_main'])

top_n = 10
top_clubs = df[club_col].value_counts().head(top_n).index
df_filtered = df[df[club_col].isin(top_clubs)]

grouped = df_filtered.groupby([club_col, 'position_main']).size().reset_index(name='count')

pivot_df = grouped.pivot(index=club_col, columns='position_main', values='count').fillna(0)

fig = px.bar(
    pivot_df,
    x=pivot_df.index,
    y=pivot_df.columns,
    title="Players Distribution by Position per Club (FIFA 23)"
)

fig.update_layout(
    barmode='stack',
    xaxis_title="Club",
    yaxis_title="Number of Players",
    legend_title="Position",
    xaxis_tickangle=45,
    height=600
)

fig.show()
```
