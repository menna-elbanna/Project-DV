```python
import pandas as pd
df = pd.read_csv("data/cleaned_data.csv")
df.head()
```

```python
df.columns
```

```python
top_clubs = df['club_name'].value_counts().head(10).index.tolist()
top_clubs
```

```python
filtered_df = df[df['club_name'].isin(top_clubs)]
filtered_df.shape
```

```python
selected_club = 'Liverpool'
min_age = 20
max_age = 30
temp_df = filtered_df[(filtered_df['club_name'] == selected_club) &(filtered_df['age'] >= min_age) &
(filtered_df['age'] <= max_age)]
temp_df.shape
```

```python
import plotly.express as px
scatter_fig = px.scatter(temp_df,x='age',y='overall',hover_name='short_name',
    title='Liverpool Players: Age vs Overall Rating',
    color_discrete_sequence=['lightblue'])
scatter_fig.update_traces(marker=dict(line=dict(width=1, color='black')))
scatter_fig.update_layout(xaxis_title='Age',yaxis_title='Overall Rating',template='plotly_white')
scatter_fig.show()
```

```python
bubble_df = temp_df.dropna(subset=['value_eur'])
bubble_df = bubble_df[bubble_df['value_eur'] > 0]
bubble_fig = px.scatter(bubble_df,x='age',y='potential',size='value_eur',hover_name='short_name',
    title='Liverpool Players: Age vs Potential (Bubble Size = Value)',
    color_discrete_sequence=['lightblue'],size_max=50)
bubble_fig.update_traces(marker=dict(line=dict(width=1, color='black')))
bubble_fig.update_layout(xaxis_title='Age',yaxis_title='Potential',template='plotly_white')
bubble_fig.show()
```
